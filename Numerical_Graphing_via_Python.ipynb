```python
# Numerical Analysis Practice

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy import optimize

# Set up matplotlib for inline display in Colab
%matplotlib inline

# 1. Basic Array Operations and Plotting
def array_operations():
    # Create sample arrays
    x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
    y1 = np.sin(x)
    y2 = np.cos(x)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y1, 'b-', label='Sine')
    plt.plot(x, y2, 'r-', label='Cosine')
    plt.title('Basic Trigonometric Functions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

# 2. Numerical Integration Example
def numerical_integration():
    # Define a function to integrate
    def f(x):
        return x**2 * np.exp(-x)

    # Perform integration
    result, error = integrate.quad(f, 0, np.inf)
    print(f"Numerical Integration Result: {result}")
    print(f"Estimated Error: {error}")

    # Plot the function
    x = np.linspace(0, 10, 200)
    y = f(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'g-')
    plt.title('Function: x²e⁻ˣ')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()

# 3. Root Finding Example
def root_finding():
    # Define a function
    def f(x):
        return x**3 - 2*x - 5

    # Find root using different methods
    root_newton = optimize.newton(f, 2)
    root_bisect = optimize.bisect(f, 1, 3)

    print(f"Root (Newton's method): {root_newton}")
    print(f"Root (Bisection method): {root_bisect}")

    # Plot the function and root
    x = np.linspace(1, 3, 100)
    y = f(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'm-')
    plt.plot(root_newton, f(root_newton), 'ro', label='Root')
    plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
    plt.title('Root Finding: x³ - 2x - 5')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

# 4. Numerical Differentiation Example
def numerical_differentiation():
    # Define a function
    def f(x):
        return np.sin(x)

    # Analytical derivative
    def df_exact(x):
        return np.cos(x)

    # Numerical derivative
    h = 1e-6
    x = np.linspace(0, 2*np.pi, 100)
    df_numeric = (f(x + h) - f(x)) / h

    # Plot comparison
    plt.figure(figsize=(10, 6))
    plt.plot(x, df_exact(x), 'b-', label='Exact Derivative')
    plt.plot(x, df_numeric, 'r--', label='Numerical Derivative')
    plt.title('Numerical vs Exact Differentiation')
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run all examples
def main():
    print("1. Array Operations and Plotting")
    array_operations()
    print("\n2. Numerical Integration")
    numerical_integration()
    print("\n3. Root Finding")
    root_finding()
    print("\n4. Numerical Differentiation")
    numerical_differentiation()

# Execute the code
if __name__ == "__main__":
    main()
```
