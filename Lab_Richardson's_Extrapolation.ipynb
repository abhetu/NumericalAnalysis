#Richadrson's Extrapolation Lab


```python
# Question1: Numerical Derivative Estimation
import numpy as np

def central_diff(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

# a. f(x) = cos(x) at x = 0
f1 = np.cos
x1 = 0
print("a. f'(x) = d/dx cos(x) at x=0 ≈", central_diff(f1, x1))  # Expect ~0

# b. f(x) = arctan(x) at x = 1
f2 = np.arctan
x2 = 1
print("b. f'(x) = d/dx arctan(x) at x=1 ≈", central_diff(f2, x2))  # Expect ~1/2

# c. f(x) = |x| at x = 0
f3 = lambda x: abs(x)
x3 = 0
print("c. f'(x) = d/dx |x| at x=0 ≈", central_diff(f3, x3))  # Not differentiable
```

```python
#Question no. 4
import matplotlib.pyplot as plt

# Function and true derivative
f = np.sin
df_true = np.cos(np.pi / 4)
x = np.pi / 4

# Central difference
def D(h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Richardson's extrapolation
def D_rich(h):
    return (4 * D(h / 2) - D(h)) / 3

# Step sizes: h = 4^-n
hs = [4**(-n) for n in range(13)]
errors_basic = [abs(D(h) - df_true) for h in hs]
errors_rich = [abs(D_rich(h) - df_true) for h in hs]

# Plotting
plt.plot(hs, errors_basic, label="Central Difference Error")
plt.plot(hs, errors_rich, label="Richardson Extrapolation Error")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("h")
plt.ylabel("Absolute Error")
plt.legend()
plt.title("Error vs h (log-log scale)")
plt.grid(True)
plt.show()
```
