```python
# Question no. 4 Newton's Method for solving 2x(1 - x^2 + x)ln(x) = x^2 - 1 in [0,1]
import numpy as np

nmax = 1000
x = 0.5  # Initial guess (must be > 0)
TOL = 10**(-6)

for i in range(1, nmax):
    if x <= 1e-6:  # Prevent log(0) issues by setting a lower bound
        x = 1e-6  # Reset to a small positive value to avoid crashing

    fx = 2*x*(1 - x**2 + x)*np.log(x) - (x**2 - 1)
    fp = 2*(1 - x**2 + x)*np.log(x) + (2*x*(1 - x**2 + x)/x) - 2*x

    if abs(fp) < 1e-10:  # Prevent division by zero
        print("Derivative too small, stopping.")
        break

    d = fx / fp
    x_new = x - d

    # Ensure x does not go negative or too close to zero
    if x_new <= 1e-6:
        x_new = 1e-6

    x = x_new

    if abs(d) < TOL:
        break

print(x)
print(i)
```

```python
# Question 8: Newton's Method for solving x^3 - x^2 - x - 1 = 0
import numpy as np

# Define the function f(x) = x^3 - x^2 - x - 1
def f(x):
    return x**3 - x**2 - x - 1

# Define its derivative f'(x)
def df(x):
    return 3*x**2 - 2*x - 1

# Newton's Method Implementation
def newton_method(f, df, x, tol=1e-6, nmax=100):
    for i in range(1, nmax):
        fx = f(x)
        fp = df(x)

        if abs(fp) < 1e-10:  # Prevent division by zero
            print("Derivative too small, stopping.")
            break

        d = fx / fp
        x = x - d

        print(f"Iteration {i}: x = {x:.6f}, f(x) = {fx:.6e}")

        if abs(d) < tol:
            break
    return x

# Choose a suitable initial guess
x0 = 1.5  # Based on rough estimation

# Solve the equation
print("\nSolving x^3 - x^2 - x - 1 = 0 using Newton's Method\n")
root = newton_method(f, df, x0)

print(f"\nEstimated Root: {root:.6f}")
```

```python
import math

def newton(f, f_prime, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        f_primex = f_prime(x)
        if abs(fx) < tol:
            return x, i
        if f_primex == 0:  # Prevent division by zero
            return None, i  # Indicate failure
        x = x - fx / f_primex
        print(f"Iter {i+1}: x = {x:.6f}, f(x) = {fx:.6f}") # Print iterate and f(x)
    return None, max_iter  # Indicate failure (no convergence)

# Example 1: f(t) = tan(t) - t
def f1(t):
    return math.tan(t) - t

def f1_prime(t):
    return 1 / math.cos(t)**2 - 1

x0_1 = 7.0

# Example 2: g(t) = e^t - sqrt(t + 9)
def g1(t):
    return math.exp(t) - math.sqrt(t + 9)

def g1_prime(t):
    return math.exp(t) - 1 / (2 * math.sqrt(t + 9))

x0_2 = 2.0


# Run and print results
print("Example 1:")
root1, iterations1 = newton(f1, f1_prime, x0_1)
if root1 is not None:
    print(f"Root: {root1:.6f}, Iterations: {iterations1}")
else:
    print("No convergence.")

print("\nExample 2:")
root2, iterations2 = newton(g1, g1_prime, x0_2)
if root2 is not None:
    print(f"Root: {root2:.6f}, Iterations: {iterations2}")
else:
    print("No convergence.")
```

```python
# Question no. 6, Newton's Method vs Secant Method Comparison

import numpy as np

# Newton's Method
def newton_method(f, df, x, tol=1e-6, nmax=100):
    for i in range(1, nmax):
        fx = f(x)
        fp = df(x)

        if abs(fp) < 1e-10:  # Prevent division by zero
            print("Derivative too small, stopping.")
            break

        d = fx / fp
        x = x - d

        print(f"Newton Iter {i}: x = {x:.6f}, f(x) = {fx:.6e}")

        if abs(d) < tol:
            break
    return x

# Secant Method
def secant_method(f, x0, x1, tol=1e-6, nmax=100):
    fx0 = f(x0)
    fx1 = f(x1)

    for i in range(1, nmax):
        d = (fx1 - fx0) / (x1 - x0)
        x2 = x1 - fx1 / d
        fx2 = f(x2)

        print(f"Secant Iter {i}: x = {x2:.6f}, f(x) = {fx2:.6e}")

        if abs(x2 - x1) < tol:
            break

        x0, x1 = x1, x2
        fx0, fx1 = fx1, fx2

    return x2

# Function 1: f(x) = x^3 - 3x + 1
def f1(x): return x**3 - 3*x + 1
def df1(x): return 3*x**2 - 3  # Derivative of f1

# Function 2: g(x) = x^3 - 2sin(x)
def f2(x): return x**3 - 2*np.sin(x)
def df2(x): return 3*x**2 - 2*np.cos(x)  # Derivative of f2

# Solve function 1
print("\nSolving f(x) = x^3 - 3x + 1 with Newton's Method\n")
x1_newton = newton_method(f1, df1, x=2)  # Use x0 = 2

print("\nSolving f(x) = x^3 - 3x + 1 with Secant Method\n")
root1_secant = secant_method(f1, x0=2, x1=x1_newton)  # Use Newton's x1 for secant method

# Solve function 2
print("\nSolving g(x) = x^3 - 2sin(x) with Newton's Method\n")
x1_newton_2 = newton_method(f2, df2, x=0.5)  # Use x0 = 0.5

print("\nSolving g(x) = x^3 - 2sin(x) with Secant Method\n")
root2_secant = secant_method(f2, x0=0.5, x1=x1_newton_2)  # Use Newton's x1 for secant method

print(f"\nFinal Roots:\nf1(x) Root: {root1_secant:.6f}\ng(x) Root: {root2_secant:.6f}")
```
